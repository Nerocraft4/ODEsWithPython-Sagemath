```python
eq1=(-13*x**2/188+3*x*y/8+399*x/752-y**2/4-315*y/752-63/376)*x
eq2=(-12/47*x^2 + 19/32*x*y + 7/94*x-y**2/4+77/752*y-399/1504)*y
sol=solve([eq1,eq2],x,y); 
show(sol)
```

```python
sol[1][0].rhs().n()#2ª solució a (0.32998,0)
```

```python
sol[2][0].rhs().n()#3ª solució a (7.34310,0)
```

```python
#Tenim 4 solucions: 
s1= [0,0]
s2= [0.32998,0]
s3= [7.34310,0]
s4= [1,1]
```

```python
#vista general del sistema
plot=plot_vector_field((eq1,eq2), (x,0,9), (y,0,1.5));
show(plot)
```

```python
#vista ampliada del sistema
plot=plot_vector_field((eq1,eq2), (x,0,2), (y,0,1.5));
show(plot)
```

```python
b=0.2
#visualitzem el que succeeix a les solucions individuals
plot1=plot_vector_field((eq1,eq2), (x,s1[0]-b,s1[0]+b), (y,s1[1]-b,s1[1]+b));
plot2=plot_vector_field((eq1,eq2), (x,s2[0]-b,s2[0]+b), (y,s2[1]-b,s2[1]+b));
plot3=plot_vector_field((eq1,eq2), (x,s3[0]-b,s3[0]+b), (y,s3[1]-b,s3[1]+b));
plot4=plot_vector_field((eq1,eq2), (x,s4[0]-b,s4[0]+b), (y,s4[1]-b,s4[1]+b));
k=graphics_array([[plot1,plot2],[plot3,plot4]]);
k.show(figsize=[8, 6])
```

```python
A6=matrix([[diff(eq1,x),diff(eq1,y)],[diff(eq2,x),diff(eq2,y)]])
show(A6)
```

```python
#Primer punt
A61=A6.subs(x=s1[0],y=s1[1]);show(A61)
A61.eigenvalues()
#Atractor, vaps negatius
#Equilibri de no participació
```

```python
#Segon punt
A62=A6.subs(x=s2[0],y=s2[1]);show(A62)
A62.eigenvalues()
#Sella, un vap positiu i un negatiu
#Punt de les preses que, si el superen, creixerà la població
#i si queden per sota, hi haurà extinció. Modela problemes
#com l'endogàmia o la falta d'organització
```

```python
#Tercer punt
A63=A6.subs(x=s3[0],y=s3[1]);show(A63)
A63.eigenvalues()
#Atractor, els vaps són negatius
#Equilibri de preses sense depredadors que no comporta 
#l'extinció de les preses (però si la dels depredadors) 
```

```python
#Quart punt
A64=A6.subs(x=s4[0],y=s4[1]);show(A64)
A64.eigenvalues()
#Repulsor, la part real és positiva
#Punt inestable (físicament impossible arribar-hi) on
#es mantindrien constants les dues poblacions
```
