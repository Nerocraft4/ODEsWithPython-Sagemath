```python
A3=matrix([[diff(eq1,x),diff(eq1,y)],[diff(eq2,x),diff(eq2,y)]]).subs(x=0,y=0);
show(A3)
A3.eigenvalues()
#Com que un eigenvalue √©s zero, no podem determinar quin tipus de node √©s amb la teoria
#No obstant, amb el retrat de fase observem que es tracta d'una sella
```
