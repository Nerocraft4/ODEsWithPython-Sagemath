```python
var('x y');
eq1=-y + 2*x*y + y**2
eq2=x + x**2 + 2*x*y - y**2
b=0.3
sol=solve([eq1,eq2],x,y);
```

```python
plot=plot_vector_field((eq1,eq2), (x,-3,1), (y,-3,1.5));
show(plot)#visio general
```

```python
#visualitzem el que succeeix a les solucions
plot1=plot_vector_field((eq1,eq2), (x,sol[0][0].rhs()-b,sol[0][0].rhs()+b), (y,sol[0][1].rhs()-b,sol[0][1].rhs()+b));
plot2=plot_vector_field((eq1,eq2), (x,sol[1][0].rhs()-b,sol[1][0].rhs()+b), (y,sol[1][1].rhs()-b,sol[1][1].rhs()+b));
plot3=plot_vector_field((eq1,eq2), (x,sol[2][0].rhs()-b,sol[2][0].rhs()+b), (y,sol[2][1].rhs()-b,sol[2][1].rhs()+b));
plot4=plot_vector_field((eq1,eq2), (x,sol[3][0].rhs()-b,sol[3][0].rhs()+b), (y,sol[3][1].rhs()-b,sol[3][1].rhs()+b));
k=graphics_array([[plot1,plot2],[plot3,plot4]]);
k.show(figsize=[8, 5])
```

```python
A2=matrix([[diff(eq1,x),diff(eq1,y)],[diff(eq2,x),diff(eq2,y)]]);
show(A2)
```

```python
#primera soluci
A21=A2.subs(x=sol[0][0].rhs(),y=sol[0][1].rhs());
show(A21)
A21.eigenvalues()
#tenen signes diferents, tenim una sella
```

```python
#segona soluci
A22=A2.subs(x=sol[1][0].rhs(),y=sol[1][1].rhs());
show(A21)
A22.eigenvalues()
#en aquest cas sn centres (purament imaginaris)
```

```python
#tercera soluci
A23=A2.subs(x=sol[2][0].rhs(),y=sol[2][0].rhs());
show(A23)
A23.eigenvalues()
#tenen signes oposats -> trobem un punt de sella
```

```python
#quarta soluci
A24=A2.subs(x=sol[3][0].rhs(),y=sol[3][0].rhs());
show(A24)
A24.eigenvalues()
#tenen signes oposats -> trobem un punt de sella
```
