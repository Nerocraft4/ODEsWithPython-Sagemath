```python
eq11=5*x+9*y-1
eq12=-x-y+2
sol=solve([eq11,eq12],x,y);
show(sol[0])
plot=plot_vector_field((eq11,eq12), (x,17/4-2,17/4+2), (y,-9/4-2,-9/4+2)); #Entorn de 17/4 i -9/4
show(plot)
```

```python
A11=matrix([[diff(eq11,x),diff(eq11,y)],[diff(eq12,x),diff(eq12,y)]]);
show(A11)
```
